```python
from graphviz import Digraph

class Node:
    def __init__(self, data, children=None):
        self.data = data
        if children is None:
            children = []
        self.children = children
        
    def add_child(self, child):
        self.children.append(child)
        
    def __repr__(self, seen=None):
        if seen is None:
            seen = set()
        if id(self) in seen:
            return f"Node(data={self.data})"
        seen.add(id(self))
        children_repr = ', '.join(child.__repr__(seen) for child in self.children)
        return f"Node(data={self.data}, children=[{children_repr}])"

def trace(root):
    # Builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v.children:
                edges.add((v, child))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    # dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'})  # LR = Left to right
    dot = Digraph('G')
    # dot = graphviz.Graph('G', filename='process.gv', engine='sfdp')

    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        # For any value in the graph, create a circular ('ellipse') node for it
        dot.node(name=uid, label="%s" % (n.data), shape='ellipse')

    for n1, n2 in edges:
        # Connect n1 to n2
        dot.edge(str(id(n1)), str(id(n2)))

    return dot
```

```python
# Creating all nodes
nodeA = Node('A')
nodeB = Node('B')
nodeC = Node('C')
nodeD = Node('D')
nodeE = Node('E')
nodeF = Node('F')
nodeG = Node('G')
nodeH = Node('H')
nodeI = Node('I')
nodeJ = Node('J')
nodeK = Node('K')

# Establishing relationships based on the graph
nodeA.add_child(nodeB)  # A -> B
nodeA.add_child(nodeC)  # A -> C
nodeA.add_child(nodeD)  # A -> D
nodeB.add_child(nodeE)  # B -> E
nodeB.add_child(nodeF)  # B -> F
nodeD.add_child(nodeG)  # D -> G
nodeD.add_child(nodeH)  # D -> H
nodeF.add_child(nodeI)  # F -> I
nodeC.add_child(nodeG)  # C -> G
nodeJ.add_child(nodeK)  # J -> K
nodeG.add_child(nodeK)
nodeF.add_child(nodeJ)

# Print nodes to see the structure
draw_dot(nodeA)
```

```python
def topo_sort(node): 
    topo = []
    visited = set()
    def build_topo(v):
        if v not in visited:
            visited.add(v)
            for child in v.children:
                build_topo(child)
            topo.append(v)
    build_topo(node)
    return topo

    
```

```python
topo_sort(nodeA)
```

```python
root = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4) 

root.add_child(node2)
root.add_child(node3)
root.add_child(node4)
root.add_child(nodeA)
node2.add_child(nodeA)


draw_dot(root)
```
